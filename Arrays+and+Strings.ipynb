```python
import string # learn more: https://python.org/pypi/strings
from random import randint
letterTimes = 0
indexLongWord = 0
#Task 1 Lists
names = ['Ana','Ben','Isabell','John','Emily','Joshua']
lenght = len(names)-1
for i in range(len(names)):
  if i % 2 == 0:
    print(names[i])
#Task 2 Lists    
print('')
print(names[randint(0,lenght)])
print('')
#Task 3 Lists
for i in reversed(names):
  print(i)
  
print('')
#Task 4 Lists
words = ['hello','apple','car','bus','teacher']
for i in range(len(words)):
  print(words[i][len(words[i])-1])
  
print('')
#Task 5 Lists
for i in range(len(words)):
  print(words[i][randint(0,len(words[i])-1)])
print('')
#Lists Page 2 Task 1
temperature = [16,13,12,15,14,12,11,20,12,19,12,12,17,12,12]

temperature.insert(temperature.index(20)+1,5)
print(temperature)
print('')
#Lists Page 2 Task 2
for i in range(3):
  temperature.append(7)
print(temperature)
print('')
#Lists Page 2 Task 3
temperature.remove(11)
print(temperature)
print('')
#Lists Page 2 Task 4
temperature[temperature.index(17)] = 16
temperature[temperature.index(16)] = 17
print(temperature)
print('')
#Lists Page 2 Task 5
while 12 in temperature:
  temperature.remove(12)
print(temperature)
print('')
#This is where the Strings part begins
longWord = 'PSEUDOPSEUDOHYPOPARATHYROIDISM'
longWordLetter = string.ascii_uppercase[randint(0,25)]
for i in range(len(longWord)):
  if longWordLetter == longWord[i]:
    letterTimes = letterTimes + 1
if longWordLetter not in longWord:
  print('The letter', longWordLetter, 'is NOT in the word.')
else:
  print('The letter', longWordLetter, 'is IN the word', letterTimes, 'times.')
```
